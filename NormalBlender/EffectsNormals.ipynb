```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io as io
import skimage as ski
```

```python
BasePlane = ski.io.imread("BasePlane.png")
EditedPlane = ski.io.imread("EditedPlane.png")
EditedMesh = ski.io.imread("EffectsMesh.png")
```

```python
alpha = np.tile([0,0,0,1], ((BasePlane.shape[0], BasePlane.shape[1], 1)))
```

```python
Difference = ((EditedPlane - BasePlane) / 255) + alpha
```

```python
fig, ax = plt.subplots(figsize=(17.5,5))

frame1 = plt.gca()
frame1.axes.xaxis.set_ticklabels([])
frame1.axes.yaxis.set_ticklabels([])

plt.subplot(1,3,1)
plt.title("No edits")
plt.imshow(BasePlane)

plt.subplot(1,3,2)
plt.title("With edits")
plt.imshow(EditedPlane)

plt.subplot(1,3,3)
plt.title("Difference")
plt.imshow(Difference)



plt.show()
```
